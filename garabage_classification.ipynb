```python
import tensorflow as tf

train_data = tf.keras.preprocessing.image_dataset_from_directory(
    "dataset/train",
    image_size=(128, 128),
    batch_size=32
)

model = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1./255),
    tf.keras.layers.Conv2D(16, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(len(train_data.class_names), activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(train_data, epochs=5)

model.save("waste_model.h5")
print("Model saved successfully!")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import os

# Dataset path
base_dir = "/content/drive/MyDrive/smart waste classification using AI/GARBAGE CLASSIFICATION"
# Point flow_from_directory to the 'images' subdirectories within train and test
train_images_path = os.path.join(base_dir, "train", "images")
test_images_path = os.path.join(base_dir, "test", "images")

# Image preprocessing
train_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_images_path, # Now pointing directly to the directory expected to contain class folders
    target_size=(128, 128),
    batch_size=32,
    class_mode="categorical",
)

test_datagen = ImageDataGenerator(rescale=1./255)
test_data = test_datagen.flow_from_directory(
    test_images_path, # Now pointing directly to the directory expected to contain class folders
    target_size=(128, 128),
    batch_size=32,
    class_mode="categorical"
)

# CNN Model
# Num_classes will now be inferred correctly if the data structure is proper (class subdirs under train_images_path)
# If it still finds 0 images, then the directory 'train_images_path' is empty or doesn't have class subdirs.
num_classes = len(train_data.class_names) if train_data.samples > 0 else 6 # Fallback to 6 if no samples found

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(num_classes, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
if train_data.samples == 0:
    print("\nWARNING: No training samples found. Please ensure that '" + train_images_path + "' contains subdirectories named after your classes (e.g., 'BIODEGRADABLE/', 'CARDBOARD/'), and that these subdirectories contain image files.")
else:
    print("\nStarting model training...")
    model.fit(train_data, epochs=5)

    # Test accuracy
    if test_data.samples == 0:
        print("\nWARNING: No test samples found. Cannot evaluate model accuracy.")
    else:
        loss, accuracy = model.evaluate(test_data)
        print("Accuracy:", accuracy)

    # Save model
    model.save("waste_classifier.h5")
    print("Model Saved!")
```
